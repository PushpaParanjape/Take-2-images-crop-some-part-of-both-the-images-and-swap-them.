```python
#import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Get & display 1st img
pic1 = cv2.imread('img1.jpg')
plt.imshow(pic1)
```

```python
#Display the shape of 1st img
pic1.shape
```

```python
#Get & display 2nd img
pic2 = cv2.imread('img2.jpg')
plt.imshow(pic2)
```

```python
#Display the shape of 2nd img
pic2.shape
```

```python
#Crop some part of 1st img
crop_pic1 = pic1[0:160 , 100:250]
plt.imshow(crop_pic1)
```

```python
#Crop some part of 2nd img
crop_pic2 = pic2[0:180 , 150:300]
plt.imshow(crop_pic2)
```

```python
#Display the shape of 1st cropped img 
crop_pic1.shape
```

```python
#Display the shape of 2nd cropped img 
crop_pic2.shape
```

```python
#Resize both cropped imgs
crop_pic1 = cv2.resize(crop_pic1 , (150 , 180))
crop_pic2 = cv2.resize(crop_pic2 , (150 , 160))
```

```python
#Display 1st crop img
plt.imshow(crop_pic1)
```

```python
#Display 2nd crop img
plt.imshow(crop_pic2)
```

```python
pic1[0:160 , 100:250] = crop_pic2
plt.imshow(pic1)
```

```python
pic2[0:180 , 150:300] = crop_pic1
plt.imshow(pic2)
```
